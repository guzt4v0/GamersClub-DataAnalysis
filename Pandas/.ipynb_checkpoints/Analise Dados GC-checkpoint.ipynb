```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Logo após a importação da biblioteca, vamos carregar o dataset que será utilizado.

Esse dataset foi cedido gentilmente pela equipe da GamersClub, o dataset contém informações de partidas jogadas do popular game Counter-Strike Global Offensive (CS-GO).

```python
df = pd.read_csv("data/tb_lobby_stats_player.csv", sep=",")
```

Vamos usar a função **.head()** para ter um overview do dataset que estamos lidando, suas linhas e colunas, limitaremos a visualização nos 10 primeiros registros para não precisar carregar todo o dataset no momento. 

```python
df.head(10)
```

Podemos citar algumas colunas interessantes para nossa análise como:
* __idLobby__ - Identificador da partida que foi registrada
* __idPlayer__ - Identificador do jogador
* __qtKill__ - Quantidade de abates por jogador
* __qtDeath__ - Quantidade de mortes por jogador
* __vlLevel__ - Level do jogador
* __descMapName__ - Mapa que foi jogado
* __qtHitHeadshot, qtHitChest, qtHitStomach, qtHitLeftAtm, qtHitRightArm, qtHitLeftLeg, qtHitRightLeg__ - Partes do corpo que foram acertadas para o abate por quantidade

Vamos verificar a quantidade de jogadores __(idPlayer)__ que tem nesse dataset, lembrando que tem jogadores que se repetem então será necessário tirar os duplicados com a função __.drop_duplicates()__

## Número de Players

```python
players = df["idPlayer"]
players.drop_duplicates(inplace=True)
players.count()
```

Ao utilizar a função __.count()__ vimos que a quatidade de jogadores únicos no dataset são de __2469 jogadores__.

-------------------------------------------------

Com esse bom número de jogadores vamos verificar os mapas __(descMapName)__ mais jogados por eles em porcentagem.

## Mapas mais jogados

```python
df["descMapName"].value_counts(normalize=True).round(2)
```

Com isso vimos que o mapa mais jogado é __Mirage__ e o menos jogado é __Ancient__, mas vamos tentar deixar um pouco mais visual essa análise, com isso vamos usar de ajuda a biblioteca __Seaborn__ e um dos seus tipos de gráficos mais populares o __catplot()__, que torna a plotagem mais fácil de variáveis categóricas.

```python
figMap = sns.catplot(y='descMapName', kind='count', data=df,order=df['descMapName'].value_counts().index)
figMap.set_xlabels('Frequência', fontsize=15)
figMap.set_ylabels('Mapas', fontsize=15)
```

O que está por trás dessa preferência de mapas?
Eu como jogador do jogo posso dizer alguns fatores como __Complexidade do Mapa__ (Alguns mapas são complexos de se jogar devido a arquitetura dele, como exemplo __nuke__), __Peso do Mapa__ (Mirage é um dos mapas mais leves de se jogar, assim computadores menos potentes conseguem ter uma boa performance), __Nostalgia__ (Por ser um jogo que tem mais de 20 anos, considerando as outras versões, alguns mapas tem desde o começo, assim a familiaridade dos nomes leva a escolha do mapa).

-----------------------------------

Aproveitando que estamos falando de mapa, vamos conferir a distribuição dos mapas mais jogados por Nivel __(vlLevel)__, lembrando que a GC possui um ranking de habilidade que vai de __0__ ao nivel __21__, sendo que 21 o mais alto nivel. Usaremos o __(idLobbyGame)__ para contar a quantidade de partida em determinado mapa.

## Porcentagem de Mapas Jogados por Nivel

```python
#Vamos criar um DataFrame com os niveis, mapas e o identificador da partida para facilitar o processo de transformação
dfGroupMap = df[["vlLevel", "descMapName", "idLobbyGame"]]
dfGroupMap = pd.DataFrame(data=dfGroupMap)
dfGroupMap = dfGroupMap.set_index("descMapName")
dfGroupMap
```

```python
#Aqui iremos agrupar o dataset por nivel e por mapa, o identificador da partida iremos contar e ver como fica distribuido entre o agrupamento
gMap = dfGroupMap.groupby(["descMapName", "vlLevel"]).count()
gMap = pd.DataFrame(data=gMap)

#Em seguida iremos colocar os mapas por coluna para facilitar a visualização
dfMapLvl = gMap.pivot_table('idLobbyGame', ["vlLevel"], 'descMapName')
dfMapLvl
```

```python
#Vamos usar uma pequena formula para poder visualizar o resultado distribuido em porcentagem de cada mapa dentro de 100%
stackedMapLvl = dfMapLvl.apply(lambda x: x*100/sum(x), axis=1)

#Usaremos o gráfico em empilhado em barras para facilitar a visão sobre a distribuição dos mapas por nivel, e não precisar gerar varios subplots
stackedMapLvl.plot(kind='bar', stacked=True, figsize=(20,10)).legend(loc='upper center', ncol=3, title="Porcentagem de Mapas Mais Jogados por Nivel")
plt.xlabel("Nivel dos Jogadores")
plt.ylabel("Porcentagem dos Mapas")
```

Como podemos ver nessa distribuição os resultados batem com os mapas mais jogados no geral, mas podemos ver algumas coisas interessantes, como os __niveis 21__ jogam menos __inferno__ e mais __mirage__ que os outros jogadores que mantém uma certa regularidade, e também é os que mais jogam __vertigo__.  

Foi necessário fazer esse tipo de gráfico pois um __único de barras simples__ não seria muito justo, pois não tem o mesmo número de dados de partida por nivel, e se gerasse um gráfico pra cada nivel ou mapa não seria muito bom de se __visualizar__.

----------------------------------

Agora vamos conferir a média de __headshot = tiro na cabeça (qtHs)__ por partida, dividido por nivel de jogador, vamos descobrir qual nivel se sai melhor nessa comparação.

## Média de HS/Partida por NVL

```python
meanHS = df.groupby(["vlLevel"]).agg({"qtHs": "mean"})
meanHS
```

```python
#Um gráfico simples de linha já vai suprir nossa nessecidade
meanHS.plot(figsize=(20,5))
plt.legend(['Média de HS por Partida'])
plt.xlabel("Nivel do Jogador")
plt.ylabel("Média de HS")
```

Podemos ver algo interessante aqui, como o __nivel 0__ acerta mais __HS__ que jogadores até __nivel 8__, isso é curioso pois tirando o __nivel 0__ o gráfico esta como uma escada natural, normalmente os jogadores com melhores __niveis__ tendem a ter melhor mira.

Mas ai que encontramos o motivo, jogadores de __nivel 0__ são contas novas na __Gamersclub__ e ainda não receberam o __balanceamento__ de __nivel__ então pode muito bem um jogador experiente estar jogando estando no __nivel 0__, o que explica a discrepância.

----------------------

Agora vamos ver as partes do corpo mais acertadas por nivel.

## Partes do Corpo Mais Acertadas por Nivel

```python
#Criando um dataframe com os niveis e partes do corpo acertadas
lvHit = df[["vlLevel", "qtHitHeadshot", "qtHitChest", "qtHitStomach", "qtHitLeftAtm", "qtHitRightArm", "qtHitLeftLeg", "qtHitRightLeg"]]
lvHit = lvHit.set_index("vlLevel")
lvHit
```

```python
#Vamos plotar o gráfico agrupado
groupLvHit = lvHit.groupby("vlLevel").sum()
groupLvHit.plot(figsize=(20,5))
plt.xlabel("Nivel do Jogador")
plt.ylabel("Quantidade")
```

Mesmo dividindo por nivel, percebemos que não importa o nivel, as partes mais acertadas são: __peito__, __estomago__ e __cabeça__, na respectiva ordem, alterando apenas a quantidade de dados disponiveis igualmente pra cada nivel, mas já da pra ter uma noção.



------------

Agora vamos pra reta final, vamos calcular a média do __K/D (Quantidade de abates(qtKill) dividido pela quantidade de mortes(qtDeath))__, é uma das métricas mais importantes para ver o desempenho do jogador.

## Média de K/D por NVL

```python
#Vamos criar uma tabela com o nivel, abates e mortes
kdTabel = df[['vlLevel', 'qtKill', 'qtDeath']]
kdTabel = kdTabel.set_index("vlLevel")
kdTabel
```

```python
#Vamos agrupar por nivel agora
kdGroup = kdTabel.groupby(['vlLevel']).mean().round(2)
kdGroup
```

```python
#Dividiremos o abate pela quantidade de mortes
meanKD = kdGroup.assign(KD = lambda x: (x['qtKill'] / x['qtDeath'])).round(2)
meanKD
```

```python
meanKD[["KD"]].plot(kind="bar", figsize=(20,5), colormap='Accent')
plt.xlabel("Nivel do Jogador")
plt.ylabel("Média de HS")
```

Com isso podemos ver aquela escala natural de habilidade, mas voltando aquela questão do __nivel 0__ ser discrepante.

Lembrando que quanto maior que 1.0 melhor, pois significa que __o jogador está mais matando do que morrendo__, o que no jogo em questão é um indicador muito bom de habilidade.

Para efeito de curiosidade, eu Gustavo, no mês de fevereiro de 2022 terminei com o __KD de 1.01__, e atualmente estou no __nivel 11__, podemos ver que estou acima da média em relação aos jogadores do meu nivel.
